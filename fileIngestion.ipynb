```python
with open("config.yaml", "r") as yamlfile:
    config = yaml.load(yamlfile)
    print("Read successful")
print(config)
```

# Read using config file

```python
pandass = pd.read_csv(config[0]['read']['file_name'], delimiter = config[0]['read']['delimiter'], names = config[0]['read']['columns'])
pandass.head()
```

# Number of columns validation

```python
len(pandass.columns)== len(config[0]['read']['columns'])
```

# Column name validation

```python
all( x in config[0]['read']['columns'] for x in pandass.columns)
```

# Write with pipe delimiter

```python
pandass.to_csv(config[0]['write']['file_name'], sep = config[0]['write']['delimiter'],
          compression=config[0]['write']['compression']) 
```
